```python
# Step 1 : Import req libs
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# Step 2 : Load the dataset
df = pd.read_csv("linear_regression_datasest.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Checking for outliers 

sns.boxplot(data = df)
plt.show()
```

```python
def plotFeatures(X , Y , df):
    sns.scatterplot(x=X , y = Y , data=df)
    plt.show()
```

```python

plotFeatures('Feature_1' , 'Target' , df)
plotFeatures('Feature_2' , 'Target' , df)
```

```python
# Step 6 : Dividing the data into X and y
X = df.iloc[: , :-1]
y = df.iloc[: , -1] 
# we can also select df by droping the target col for x and selecting the target col for y
```

```python
# Step 7 : Spliting the data into train and test data
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split( X , y , test_size=0.2)
```

```python
X_train.shape , y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr_preds = lr.predict(X_test)

r2_score(lr_preds, y_test)
```

```python
from sklearn.metrics import mean_absolute_error

print(mean_absolute_error(y_test , lr_preds))
```

```python
lr.coef_ , lr.intercept_
```

```python
plt.scatter(y_test, lr_preds)
plt.plot(X_test['Feature_1'] , lr_preds )
```
