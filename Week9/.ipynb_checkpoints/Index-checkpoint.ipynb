```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report , confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("decision_tree_dataset.csv")
df.head()
```

```python
df = pd.get_dummies(df , columns=['city'] ,drop_first=True)
```

```python
df.gender.replace({'Male': 1 , 'Female' : 0} , inplace=True)
```

```python
q1 = df['income'].quantile(0.80)
q2 = df['income'].quantile(0.20)
df['income'] = df['income'].clip(lower=q2 , upper=q1)
```

```python
def model(df):
    X = df.iloc[: , :-1]
    y = df.iloc[: , -1]
    X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2)
    sc = StandardScaler()
    X_train_scaled = sc.fit_transform(X_train)
    X_test_scaled = sc.transform (X_test)
    lg = DecisionTreeClassifier()
    lg.fit(X_train_scaled , y_train)
    lg_preds = lg.predict(X_test_scaled)
    print(classification_report(y_test , lg_preds))
    cm = confusion_matrix(y_test , lg_preds)
    print(cm)
```

```python
model(df)
```
