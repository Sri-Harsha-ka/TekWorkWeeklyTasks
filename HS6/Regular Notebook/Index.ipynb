```python
# DBSCAN 

# This create a groups based on density not based on distance to center .
# It does not need hyperparameter , it is not sensity to outliers , its the most cool sht
# Imp stuff in DBScan 

# there are 3 types of groups : 
# 1. Core points 
# 2. border points 
# 3. noise points
```

```python
import pandas as pd 
data = {
    "X" : [1,2,2,3,8,8,9,25,26,26,27,100],
    "Y" : [1,2,3,2,8,9,8,25,26,27,26,100]
}
data = pd.DataFrame(data)
data.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['X'] , data['Y'])
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data = sc.fit_transform(data)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score 
model = DBSCAN(eps = 3, min_samples=2)
model.fit(data)
silhouette_score(data , model.labels_)
```
