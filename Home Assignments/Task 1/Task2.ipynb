```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./dataset/Employee_Dataset.csv")
df.head()
```

```python
df['joining_date'] = pd.to_datetime(df['joining_date'], errors='coerce' , format="mixed")

failed_count = df['joining_date'].isna().sum()
failed_count
```

```python
df['joining_date'].unique()
```

```python
df['employee_id'] = (
    df['employee_id']
    .astype(str)
    .str.strip()
    .str.upper()
)

duplicate_count = df['employee_id'].duplicated().sum()
duplicate_count
```

```python
df['employee_id'].head()
```

```python
df['department'] = df['department'].astype(str).str.strip().str.upper()

df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

avg_salary = df.groupby('department')['salary'].mean()
avg_salary
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')

result = df[df['salary'].notna() & df['age'].isna()]

result.head()
```

```python
df['salary'] = pd.to_numeric(df['salary'], errors='coerce')

Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['salary'] < lower) | (df['salary'] > upper)]
outliers.head()
```

```python
df['performance_rating'] = pd.to_numeric(df['performance_rating'], errors='coerce')

median_rating = df.groupby('designation')['performance_rating'].median()
median_rating
```
