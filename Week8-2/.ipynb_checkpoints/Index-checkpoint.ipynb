```python
Q1 = df['income'].quantile(0.25)
Q3 = df['income'].quantile(0.75)
IQR = Q3 - Q1 
lb = Q1 - 1.5*IQR
up = Q3 + 1.5*IQR
df = df[(df['income'] >= lb ) & (df['income'] <=up)]
```

```python
df.head()
```

```python
# Data Preprocessing

# Handling the missing values

# income                   49  -- medium
# credit_score             50  -- 
# avg_purchase_value       50


df['income'].fillna(df['income'].median() , inplace=True)
df['credit_score'].fillna(df['credit_score'].median() , inplace=True)
df['avg_purchase_value'].fillna(df['avg_purchase_value'].median() , inplace=True)
```

```python

# Handling the Catograrical values

# gender
# ['Male' 'Female']

# region
# ['South' 'East' 'North' 'West']

df['gender'].replace({'Male' : 1 , 'Female':0} , inplace=True)
```

```python

df = pd.get_dummies(df , columns=['region'] , drop_first=True)
```

```python
df.head()
```

```python
def model(df):
    X = df.iloc[: , :-1]
    y = df.iloc[: , -1]
    X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2)
    sc = StandardScaler()
    X_train_scaled = sc.fit_transform(X_train)
    X_test_scaled = sc.transform (X_test)
    lg = LogisticRegression()
    lg.fit(X_train_scaled , y_train)
    lg_preds = lg.predict(X_test_scaled)
    print(classification_report(lg_preds , y_test))
```

```python
model(df)
```
