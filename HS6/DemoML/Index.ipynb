```python
# Hierichical Clustering

# This Clustering algo create cluster within other clusters 
# This calculate distance between other clusters
# We need to select the logest vertical line in the 
```

```python
import pandas as pd 
data = {
    'Height' : [170,162,168,190,192,188],
    'Weight' : [65,67,63,95,98,92],
}
data = pd.DataFrame(data)
data.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['Weight'], data['Height'])
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
data = sc.fit_transform(data)
```

```python
from sklearn.cluster import AgglomerativeClustering
ag = AgglomerativeClustering(n_clusters=2)
ag.fit(data)
```

```python
ag.labels_
```

```python
from scipy.cluster.hierarchy import dendrogram , linkage 
linked = linkage(data , method = 'ward')
plt.figure(figsize= (6,4))
dendrogram(linked)
plt.title('Dendrogram')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
sil = silhouette_score(data , ag.labels_)
sil
```

```python
import joblib

joblib.dump(ag,"model.pkl")
```
