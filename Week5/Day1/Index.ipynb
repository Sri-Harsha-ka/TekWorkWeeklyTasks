```python
# Importing the Necessary Libs in the Note Book 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # It is a low level lib. flexible
import seaborn as sns
```

```python
df = pd.read_csv("./Dataset/dataset.csv")
df.head()
```

## 1. Histogram

```python
# Figure size 
plt.figure(figsize = (5,5) )

# Histogram 
sns.histplot(df['age'], kde = True , color='black')
plt.title("Distribution of Age" )
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.xlim(0,100)
plt.show()
```

## Scatter plot

```python
# Scatter plot

# we use scatter plot when we need to find the correlation , outlier etc 

plt.figure(figsize = (4,2))

sns.scatterplot(x = 'age' , y='chol' , data = df , color='red' , marker='+')
```

```python
sns.scatterplot(x = 'trestbps' , y = 'chol' , data = df ,color = 'red' , marker='+')
plt.grid()
plt.xlim(0,300)
plt.ylim(0,600)
```

## Line Plot

```python
# Line plot 

#  This is Mostly used to find the patterns in the data 

sns.lineplot(x = 'age' , y = 'chol' , data = df , marker='.' , color='cyan')
plt.title("Age vs chol")
plt.grid(True)
plt.show()
```

## Box plot

```python
# Box plot 
# This is used to find Outliers 

plt.figure(figsize = (5,5))

sns.boxplot(x = 'age' , data=df , color='cyan' )
```

## Pie Chart

```python
# Pie Chart 

#  this is used to show the proportions for the data


sakura_palette = ["#ff9999", "#ffcccc"]
sns.set_theme(style='dark' , palette=sakura_palette)

plt.pie(df['target'].value_counts(),
        labels = ['No Disease' , 'Disease'] , 
        colors = sakura_palette,
        autopct = '%1.f%%',
       )
```

```python
sns.barplot(x = 'sex' , y = 'age' , data = df) 
```
