```python
# 4 
df['user_id'].replace("", np.nan , inplace = True)
df.user_id = df.user_id.str.upper()
```

```python
# 5 
df.replace('unknown' , np.nan , inplace=True)
```

```python
df['age'] = df['age'].astype("float64")
```

```python
df[(df['age'] > 100 ) | (df['age']<0)] = np.nan 
```

```python
df['age'].unique()
```

```python
# 6
df['salary'].replace('₹50000',50000 ,inplace= True)
```

```python
df['salary'] = df['salary'].astype("float64")
```

```python
df['salary'].unique()
```

```python
# 7
df['join_date'].unique()
```

```python
df['join_date'].replace('invalid' , np.nan , inplace=True)
```

```python
df['join_date'] = pd.to_datetime(df['join_date'] , format="mixed")
```

```python
df['join_date'].unique()
```

```python
# 8
df['joining_year'] = df['join_date'].dt.year
```

```python
df.joining_year.unique()
```

```python
#10
df['department'] = df['department'].str.strip().str.upper()
        
```

```python
df.city = df.city.str.title()
```

```python
df.city.unique()
```

```python
# 11
mapvalue = {
    "excellent" : 5 ,
    "4":4,
    "3" : 3,
    "2" : 2,
    "poor": 1 ,
}
df.performance_score = df.performance_score.map(mapvalue)
```

```python
df.performance_score.unique()
```

```python
#12
df.experience_years.replace('five' , 5 , inplace=True)
```

```python
df.loc[df['experience_years'] < 0 , "experience_years"] = np.nan
```

```python
df.experience_years.unique()
```
