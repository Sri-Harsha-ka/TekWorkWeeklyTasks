```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./Datasets/Dataset_1.csv")
df.head()
```

```python
# 1 : Identity all rows where order amount is non numeric but isdetlivered is marked as delivered
df['order_amount'].unique() , df['is_delivered'].unique()
```

```python
df['order_amount'].replace("₹999","999" , inplace =True)
```

```python
df.order_amount = df.order_amount.astype("float64")
```

```python
df['order_amount'].unique()
```

```python
df['is_delivered'].replace({"yes":"True" , "no" : "False"} , inplace=True)
```

```python
df.is_delivered.unique()
```

```python
df[( df.order_amount.isna() ) & ( df.is_delivered == "True" )]
```

```python
# 2:Convert order_date to date time and find orders where delivery days is negative after valid date conversion

df.order_date.unique() , df.delivery_days.unique()
```

```python
df.order_date.replace('invalid' , np.nan , inplace=True)
```

```python
df['order_date'] = pd.to_datetime(df['order_date'] , errors='coerce' , dayfirst = True)
```

```python
df.delivery_days.replace('three' , '3',inplace = True)
```

```python
df[(df['delivery_days'].astype("float64") <0 ) & ( df['order_date'].notna() ) ]
```

```python
# 3 : standazider payment mode and calculate the percentage share of each mode per city

df.payment_mode.replace({'Debit Card':'Card' , 'Credit Card' : 'Card'} ,inplace=True)
```

```python
df.payment_mode.unique()
```

```python
df.payment_mode = df.payment_mode.str.title()
```

```python
df.city = df.city.str.title()
```

```python
df.city.unique() , df.payment_mode.unique()
```

```python
val = df.groupby(['city','payment_mode']).size()
```

```python
val / val.groupby('city').sum() *100
```

```python
# 4 : detect customers whose customer age is invalid but have order amounts above the dataset median

df.customer_age.unique() , df.order_amount.uni
```

```python
orderMd = df.order_amount.median()
```

```python
df[ ( (df['customer_age'] < 0 ) | ( df['customer_age'] >100) ) & ( df['order_amount'] > orderMd)  ]
```
