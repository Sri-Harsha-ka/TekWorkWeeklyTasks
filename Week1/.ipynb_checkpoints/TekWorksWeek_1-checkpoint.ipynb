```python
project=float(input("Enter Proj score"))
Internal=float(input("Enter Internal score"))
External=float(input("Enter External score"))
if(project >=50):
    pass
else :
    print("You have failed in project",project)
if(Internal >=50):
    pass
else :
    print("You have failed in Internal",Internal)
if(External >=50):
    pass
else :
    print("You have failed in External",External)
if project >= 50 and Internal>=50 and External>= 50:
    Total= (project*0.70)+(Internal*0.10)+(External*0.20)
    print("Your Total score is ",Total)
    if(Total >= 90):
        print("A Grade")
    elif(Total >= 70) :
        print("B Grade")
    elif(Total >= 50) :
        print("C Grade")
    else:
        print("Fail")
else :
    print("Fails")
```

## Variables

- local


```python
a= "This is a Global Variable"
def localVar():
    a="This is Local VAriable"
    print(a)
    return
print(a)
localVar()
```

```python
x=5
def Upadtion():
    x=20
    print(x)
    return
print("This is Global Var",x)
print("This is local var")
Upadtion()
    
```

```python
x = 5
print("Original Global Variable Vlaue",x)
def fun_name():
    global x
    b=69
    print(x+b)
    x= x+69
fun_name()
print("New Global Variable Vlaue",x)
```

## Q)Accept Basic Sal and give gr1oss Sal

```python
Basic=int(input("Enter Basic Salary"))
Hra= 0.0
Dra=0.0
if Basic <=10000 :
    Hra = Basic*0.67
    Dra = Basic*0.80
elif Basic <=20000:
    Hra = Basic*0.69
    Dra = Basic*0.82
elif Basic >20000:
    Hra = Basic*0.72
    Dra = Basic*0.90
Gross_Sal= Basic+Hra+Dra
print("Gross Salary is ",Gross_Sal)
        
```
