```python
def verify(pp , ip , ep): 
    if(pp >= 50 and ip>=50 and ep>=50 ):
        return 1
    return 0
```

```python
# Get the data 
project = int(input())
internal = int(input())
external = int(input())

passed = verify(project , internal , external)

if(passed):
    prjP = project*0.7
    internalP = internal*0.1
    externalP = internal*0.2
    res = prjP+ internalP + externalP 
    print(int(res))
    if(res>=90):
        print("A")
    elif(res>=70):
        print("B")
    elif(res>= 50):
        print("C")
    else:
        print("F")
    
else: 
    print("WASTED")

```
