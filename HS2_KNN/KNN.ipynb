```python
from sklearn.datasets import load_iris

irs = load_iris()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data = irs.data ,  columns=irs.feature_names)
```

```python
df['target'] = irs.target
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
def model(df):
    X = df.iloc[: , :-1]
    y = df.iloc[: , -1]
    X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2)
    sc = StandardScaler()
    X_train_scaled = sc.fit_transform(X_train)
    X_test_scaled = sc.transform (X_test)
    lg = KNeighborsClassifier(n_neighbors=5 , metric='manhattan')
    lg.fit(X_train , y_train)
    lg_preds = lg.predict(X_test)
    print(classification_report(y_test , lg_preds))
    cm = confusion_matrix(y_test , lg_preds)
    print(cm)
```

```python
model(df)
```
